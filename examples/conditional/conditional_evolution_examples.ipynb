## Conditionally Guide Search to Specific Quadrants in 2D

### Implement `Condition` Class
- a condition should be derived from `hades.es.Condition`
- and implement the `evaluate` and `sample` methods:
    - The `evaluate` method maps a genotype, `x`, fitness `f`, or meta info from the `charles_instance`, to a numerical classifier that represents that particular condition/feature
    - The `sample` method is used by `hades.es.CHARLES` to provide condition values to sample the next offspring generation via the diffusion model, i.e., these are the classifier-free guidance values used for sampling new genotypes.
- Such condition instances are then passed in a tuple of conditions to the `hades.es.CHARLES` constructor.
- In the `benchmark.run` wrapper, we also apply JSON dumps for the `hades.es.CHARLES` kwargs, thus a `to_dict` method is implemented.

```python
from condevo.es.guidance import Condition
from torch import Tensor, tensor, float32, ones

class QCondition(Condition):
    def __init__(self, target=1, ):
        Condition.__init__(self)
        self.target = target

    def evaluate(self, charles_instance, x, *args, **kwargs):
        dim = np.shape(x)[-1]
        g0 = x > 0
        rc = np.zeros(len(x))              # per default, everything is labeled "0", but this only remains for the 2nd and 4th quadrant (see below)
    
        rc[g0[:,0] & g0[:,1]] = 1.         # evaluated condition for x & y > 0 -> first quadrant is labeled "1"
        rc[(~g0[:,0]) & (~g0[:,1])] = -1.  # same for x & y < 0 -> third quadrant is labeled "-1"
        if isinstance(x, Tensor):
            return tensor(rc, device=x.device, dtype=float32)
        return rc

    def sample(self, charles_instance, num_samples):
        # sample target quadrant, with label defined in "evaluate"
        return ones(num_samples, dtype=float32) * self.target

    def to_dict(self):
        return {"target": self.target, **Condition.to_dict(self)}
```

#### Analyze Data

```python
p = np.array((1., 1.))
p = p[:] / np.linalg.norm(p)

def get_data(df, ax, color="tab:blue"):
   
    d_best = []
    d_pos = []
    d_mu = []
    d_std = []
    d_fit = []    
    d_gens = []
    for r, df_r in df.groupby("run"):
        df_r = df_r.sort_values("gen")
        d_r = np.array([s.dot(p) for s in df_r["samples"]])    
        d_pos.append(d_r)
        d_f = np.array([f for f in df_r["fitness"]])
        d_fbest = d_f.argmax(axis=1)
        d_best.append([d_ri[b] for d_ri, b in zip(d_r, d_fbest)])    
        g_r = np.array([[gi] * d_r.shape[1] for gi in df_r["gen"]])       
        d_mu.append(d_r.mean(axis=1))
        d_std.append(d_r.std(axis=1))
        d_fit.append(d_f)
        d_gens.append(g_r)
    

    # violin plot
    data = []
    all_gens = np.stack(d_gens).flatten()
    all_pos = np.stack(d_pos).flatten()
    unique_gens = np.array(sorted(np.unique(np.stack(d_gens))))
    for g in unique_gens:
        gen_pos = all_pos[all_gens == g]
        data.append(gen_pos)        
    
    parts = ax.violinplot(data, showmeans=False, showmedians=False, showextrema=False)
    ax.scatter(np.stack(d_gens).flatten() + 1, np.stack(d_pos).flatten(), 
               # s=50 * np.stack(d_fit).flatten() + 1, 
               s=15,
               facecolors=color, alpha=0.5, marker=".", edgecolors="black", linewidth=0.1)

    for pc in parts['bodies']:
        pc.set_facecolor(color)
        pc.set_alpha(0.5)
        # pc.set_edgecolor('black')

    def set_axis_style(ax, labels):
        ax.set_xticks(np.arange(1, len(labels) + 1), labels=labels)
        ax.set_xlim(0.25, len(labels) + 0.75)

    # set style for the axes
    labels = unique_gens
    set_axis_style(ax, labels)

    return d_best, d_mu, d_std, d_gens, d_pos

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, sharey=True, figsize=(10, 7))
p1_best, p1_mu, p1_std, p1_gens, p1_pos = get_data(df_p1, color="tab:red", ax=ax1)
n0_best, n0_mu, n0_std, n0_gens, n0_pos = get_data(df_n0, color="tab:olive", ax=ax2)
n1_best, n1_mu, n1_std, n1_gens, n1_pos = get_data(df_n1, color="tab:blue", ax=ax3)

x = np.linspace(-3, 3, 101)[:, None]

m = o.foo_kwargs["m"]
o.foo_kwargs["m"] = np.dot((1, 1), p)
y = np.array([o(x).numpy() for _ in p1_best])

ax1.hlines([np.dot((1, 1), p)], xmin=-0.5, xmax=len(p1_best)+1.0, color="black", linestyle="--", label="Target")
ax1.imshow(y.T, extent=(-0.5, len(p1_best)+1.0, min(x), max(x)), cmap="binary_r", vmin=-1, vmax=1, aspect=0.5)
ax1.scatter([], [], color="tab:red", label="$C_{Q1}$-Data", edgecolor="black", linewidth=0.25)
ax1.scatter([], [], color="tab:olive", label="$C_{Q0}$-Data", edgecolor="black", linewidth=0.25)
ax1.scatter([], [], color="tab:blue", label="$C_{Q3}$-Data", edgecolor="black", linewidth=0.25)
ax1.legend()

ax2.imshow(y.T, extent=(-0.5, len(p1_best)+1.0, min(x), max(x)), cmap="binary_r", vmin=-1, vmax=1, aspect=0.5)

ax3.hlines([np.dot((-1, -1), p)], xmin=-0.5, xmax=len(p1_best)+1.0, color="black", linestyle="--", label="Target")
ax3.imshow(y.T, extent=(-0.5, len(p1_best)+1.0, min(x), max(x)), cmap="binary_r", vmin=-1, vmax=1, aspect=0.5)

ax3.set_ylim([-3, 3])
ax3.set_xlabel("Generations")
ax1.set_ylabel("$\\hat x$")
ax2.set_ylabel("$\\hat x$")
ax3.set_ylabel("$\\hat x$")
        
o.foo_kwargs["m"] = m
fig.savefig("local/img/conditional_double_peak.png", transparent=True, dpi=300)
```
